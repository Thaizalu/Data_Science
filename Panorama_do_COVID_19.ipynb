# **Panorama do COVID-19**

A pandemia de COVID-19, causada pelo novo coronavírus SARS-CoV-2, começou em dezembro de 2019 na cidade de Wuhan, na China. A doença rapidamente se espalhou para outros países, levando a Organização Mundial da Saúde (OMS) a declarar uma emergência de saúde pública de preocupação internacional em 30 de janeiro de 2020 e, posteriormente, uma pandemia em 11 de março de 2020. A COVID-19 se caracteriza por uma ampla gama de sintomas, desde leves, como febre e tosse, até graves, como dificuldade respiratória e falência múltipla de órgãos.

<center><img alt="Colaboratory logo" width="50%" src="https://static.poder360.com.br/2022/12/coronavirus-imagem-1-848x477.jpg"></center>

A pandemia teve um impacto profundo na saúde global, economia e vida cotidiana, resultando em medidas extensivas de saúde pública, incluindo quarentenas, uso de máscaras, distanciamento social e campanhas de vacinação em massa. A resposta global à pandemia envolveu esforços de colaboração em pesquisa científica, desenvolvimento de vacinas e tratamentos, bem como adaptações significativas nas práticas sociais e econômicas para mitigar a propagação do vírus e salvar vidas.

Com o objetivo analisar a evolução da pandemia, esse notebook tem por objetivo observar a evolução da pandemia ao longo dos anos.

O excesso de informação das tabelas acima, atrapalham um pouco na visualização dos dados. Serão então criado um gráfico de barras, com o total de mortos pela covid-19, nas datas de 30/04/2021 e 30/04/2022,  com  o intuito  de comparar  como foi  a evolução da pandemia após 1 ano.

```
# Convertendo a coluna de datas para o tipo datetime, se ainda não estiver
df_limpo['date'] = pd.to_datetime(df_limpo['date'])

# Filtrando o DataFrame para incluir apenas os dados de 30 de abril de 2021 e 30 de abril de 2022
df_abril_2021 = df_limpo[df_limpo['date'] == '2021-04-30']
df_abril_2022 = df_limpo[df_limpo['date'] == '2022-04-30']

# Selecionando os 5 países com mais mortes em 30 de abril de 2021
top_5_total_deaths_2021 = df_abril_2021[['location', 'total_deaths']].sort_values(by="total_deaths", ascending=False).head(5)

# Selecionando os 5 países com mais mortes em 30 de abril de 2022
top_5_total_deaths_2022 = df_abril_2022[['location', 'total_deaths']].sort_values(by="total_deaths", ascending=False).head(5)

# Criando uma figura e uma grade de subplots com 1 linha e 2 colunas
fig, axs = plt.subplots(1, 2, figsize=(15, 8))

# Função para colorir as barras e destacar o Brasil
def plot_bars(ax, data, title):
    bars = ax.bar('location', 'total_deaths', data=data, color=['red' if loc == 'Brazil' else 'blue' for loc in data['location']])
    ax.set_title(title)
    ax.set_xlabel('País')
    ax.set_ylabel('Total de Mortes')
    for bar in bars:
        height = bar.get_height()
        ax.annotate('{}'.format(height),
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),
                    textcoords="offset points",
                    ha='center', va='bottom')
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_visible(False)
    ax.yaxis.set_visible(False)

# Plotando o gráfico de barras para 30 de abril de 2021
plot_bars(axs[0], top_5_total_deaths_2021, 'Top 5 países por total de mortes em 30 de Abril de 2021')

# Plotando o gráfico de barras para 30 de abril de 2022
plot_bars(axs[1], top_5_total_deaths_2022, 'Top 5 países por total de mortes em 30 de Abril de 2022')

# Ajustando o layout para evitar sobreposição de rótulos
plt.tight_layout()

# Exibindo os gráficos
plt.show()
```

Tanto em abril de 2021 quanto em abril de 2022, o Brasil figurava em segundo lugar na quantidade de mortos pela Covid-19. É interessante também notar, que os Estados Unidos, liderou no ranking em número de mortos.

* [OPCIONAL] Plotar um gráfico de dispersão considerando `gdp_per_capita` (ou seja, o PIB dos países) pelo número total de mortos. Pegue a data mais recente apenas.

```
# Convertendo a coluna de datas para o tipo datetime, se ainda não estiver
df_limpo['date'] = pd.to_datetime(df_limpo['date'])

# Filtrando o DataFrame para incluir apenas os dados de abril de 2021
df_abril_2021 = df_limpo[(df_limpo['date'] == '2021-04-30')]

# Filtrando o DataFrame para incluir apenas os dados de março de 2020 a abril de 2022
df_periodo = df_limpo[(df_limpo['date'] == '2022-04-30')]

# Criando uma figura e uma grade de subplots com 1 linha e 2 colunas
fig, axs = plt.subplots(1, 2, figsize=(15, 6))

# Função para plotar os gráficos de dispersão e destacar os países específicos
def plot_scatter(ax, df, title, countries_to_highlight):
    scatter = ax.scatter(df['gdp_per_capita'], df['total_deaths'], alpha=0.5)
    ax.set_title(title)
    ax.set_xlabel('GDP per Capita')
    ax.set_ylabel('Total de Mortes')

    # Destacar países específicos
    for country in countries_to_highlight:
        country_data = df[df['location'] == country]
        if not country_data.empty:
            ax.scatter(country_data['gdp_per_capita'], country_data['total_deaths'], label=country, s=100) # Aumenta o tamanho do ponto para destaque
            for _, row in country_data.iterrows():
                ax.annotate(country, (row['gdp_per_capita'], row['total_deaths']), textcoords="offset points", xytext=(5,5), ha='right')

# Lista de países a destacar
countries_to_highlight = ["United States", "Brazil", "India", "Mexico", "Australia"]

# Plotando o primeiro gráfico (scatter plot) - Abril de 2021
plot_scatter(axs[0], df_abril_2021, 'Relação entre GDP per Capita e Total de Mortes em 30 de Abril de 2021', countries_to_highlight)

# Plotando o segundo gráfico (scatter plot) - Abril de 2022
plot_scatter(axs[1], df_periodo, 'Relação entre GDP per Capita e Total de Mortes em 30 de Abril de 2022', countries_to_highlight)

# Ajustando o layout para evitar sobreposição de rótulos
plt.tight_layout()

# Adicionando legenda
axs[0].legend()
axs[1].legend()

# Exibindo os gráficos
plt.show()
```

### **Análise Exploratória Inicial**

Identificando o primeiro caso de COVID no Brasil.

```
# Identificar quando foi registrada o primeiro caso de covid no Brasil
df_primeiro_caso = df_brasil.loc[df_brasil['new_cases'] != 0.0].head(1)
df_primeiro_caso
```

O dataset extraido está sempre atualizando, então registrar o caso pelo número da linha, além de não obter dados em tempo real. Para isso então, foi usado um código para que a linha fosse indicada automaticamente.

```
# Extraindo a data do primeiro caso
data_primeiro_caso = df_primeiro_caso['date'].values[0]

# Convertendo a data para um formato mais legível
data_formatada = pd.to_datetime(data_primeiro_caso).strftime('%Y-%m-%d')

# Imprimindo a data
print(f'O primeiro caso foi registrado em {data_formatada}')
```

Identificando a primeira morte pela doença

```
# Identificar quando foi registrada a primeira morte de covid no Brasil
df_primeira_morte = df_brasil.loc[df_brasil['new_deaths'] != 0.0].head(1)
df_primeira_morte
```

```
# Extraindo a data da primeira morte
data_primeira_morte = df_primeira_morte['date'].values[0]

# Convertendo a data para um formato mais legível
data_formatada_morte = pd.to_datetime(data_primeira_morte).strftime('%Y-%m-%d')

# Imprimindo a data
print(f'A primeira morte foi registrada em {data_formatada_morte}')
```

Calculando o tempo que levou para o primeiro caso até a primeira morte ser notificada.

```
# Encontrar a data do primeiro caso
primeiro_caso = df_brasil.loc[df_brasil['new_cases'] != 0.0].head(1)
data_primeiro_caso = pd.to_datetime(primeiro_caso['date'].values[0])

# Encontrar a data da primeira morte
primeira_morte = df_brasil.loc[df_brasil['new_deaths'] != 0.0].head(1)
data_primeira_morte = pd.to_datetime(primeira_morte['date'].values[0])

# Calcular a diferença de dias entre as duas datas
diferenca_dias = (data_primeira_morte - data_primeiro_caso).days

# Imprimir a diferença de dias
print(f'A diferença de dias entre o primeiro caso e a primeira morte é de {diferenca_dias} dias.')
```

### **Visualização de Dados**

```
# Definindo x como a coluna de datas formatada
x = list(range(len(df_brasil['total_cases'])))  # Ajusta o tamanho de x para corresponder a y

# Extraindo os dados de casos totais e mortes totais
y = df_brasil['total_cases']
y2 = df_brasil['total_deaths']

# Criando a figura e os eixos para os gráficos
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(12, 10))

# Plotando os casos totais em escala linear
ax1.set_title('Casos no Brasil')
ax1.set_xlabel('Dias')
ax1.set_ylabel('Casos')
ax1.plot(x, y, color='blue')

# Plotando os casos totais em escala logarítmica
ax2.set_title('Casos no Brasil (Escala logarítmica)')
ax2.set_xlabel('Dias')
ax2.set_ylabel('Casos')
ax2.set_yscale('log')
ax2.plot(x, y, color='blue')

# Plotando as mortes totais em escala linear
ax3.set_title('Mortes no Brasil')
ax3.set_xlabel('Dias')
ax3.set_ylabel('Mortes')
ax3.plot(x, y2, color='red')

# Plotando as mortes totais em escala logarítmica
ax4.set_title('Mortes no Brasil (Escala logarítmica)')
ax4.set_xlabel('Dias')
ax4.set_ylabel('Mortes')
ax4.set_yscale('log')
ax4.plot(x, y2, color='red')

# Ajustando o layout para evitar sobreposição de elementos
plt.tight_layout()

# Ajustando o espaçamento vertical entre os subplots
plt.subplots_adjust(hspace=0.4)

# Exibindo os gráficos
plt.show()
```

Pelos gráficos acima, é nítido perceber que o ápice da pandemia foi o ano de 2021, onde foi percebido uma variação de 400 mil morte somente nesse ano. A estabilidade dos casos e do número de mortes, coincidiu com o avanço da vacinação no país.

## **Conclusão**



A pandemia de COVID-19 teve um impacto profundo em todo o mundo, incluindo o Brasil. A vacinação se mostrou essencial na redução de casos graves e mortes, destacando a importância de campanhas de vacinação eficazes. A falta de adoção de medidas de proteção, como uso de máscaras e higiene das mãos, contribuiu significativamente para a propagação do vírus.

O distanciamento social foi uma medida crucial para reduzir a transmissão. Países como Nova Zelândia, Austrália e Coreia do Sul adotaram políticas rigorosas de distanciamento social e controle de fronteiras, conseguindo evitar grandes números de infecções e mortes. Esses exemplos sublinham a eficácia das políticas de saúde pública na gestão de pandemias. A experiência global reafirma que medidas combinadas de vacinação, proteção individual e distanciamento social são fundamentais para controlar a disseminação de doenças infecciosas.
